# 2. Main Part

First of all I started by defining functions that would calculate the difference score petween two images. 

**'diff_score'** function would later on be used in order to compare tiles from an image to the sample tile images from the database. Compares two incoming images and using 'rms_vec' function returns the average difference score for all pixels.

**'rms_vec'** is a helping function that calculated RMS error(difference) value for two pixels. Since pixels store three values (RGB) output of this function is also a vector with three RMS scores corresponding to three chanels. It uses simple mathematical functions for calculations.

```python
def rms_vec(vec1,vec2,length):
    rms_v = np.zeros(length)
    if length>1:
        for i in range(0,length-1):
        #rms_v = pow(np.mean(vec1)-np.mean(vec2),2)
            rms_v[i] = pow(int(vec1[i])-int(vec2[i]),2)
    else:
        rms_v = pow(vec1-vec2,2)
    return rms_v

def diff_score(candidate, target, vec_length):
    diff_score_arr = np.zeros(vec_length)
    #diff_score_arr = 0.0
    y, x, _ = np.shape(candidate)
    for i in range(x):
        for j in range(y):   
            diff_score_arr = diff_score_arr + rms_vec(candidate[i][j],target[i][j],vec_length)
    diff_score = np.mean(diff_score_arr)
    return diff_score
```

After this I wrote a function that imports images from the given folder and creates a database available for later mosaicing. It also automaticaly reshapes them into a specific tile dimensions for future use.

```python
def img_base(size_x,size_y):
    images = []
    labels = np.arange(1,100)
    path = 'db_img'
    for img in os.listdir(path):
        img = cv2.imread(os.path.join(path,img))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = cv2.resize(img, (size_x,size_y), interpolation = cv2.INTER_AREA)
        images.append(img)
    return images
```

The next function iterates through all of the tile images presented in a database and calculates difference score for each of them. After that it return an image tile with the minimum difference score.

```python
def best_tile(db, img_inp, vec_length):
    score = np.zeros(np.shape(db)[0]-1)
    lx = np.shape(db[0])[0]
    ly = np.shape(db[0])[1]
    i = 0
    for i in range(0,98):
        curr_score = diff_score(db[i], img_inp, vec_length)
        score[i] = curr_score
    min_v = score[0]
    k = 0
    for a in score:
        if a <= min_v:
            min_v = a
            min_ind = k
        k=k+1
    return db[min_ind]
```

Finally **'img_mosaic'** is the function that generates a mosaic image. It iterates through pixels of an input images with a step of the tile size, and using **'best_tile'** function finds a tile image from the database that fits the best and replaces it.

```python
def img_mosaic(image,db,tile_x,tile_y,vec_length):
    new_image = image
    y, x, _ = np.shape(new_image)
    for i in range(0,y,tile_y):
        for j in range(0,x,tile_x):
            old_tile = new_image[i:i + tile_y, j:j + tile_x]
            new_tile = best_tile(db, old_tile, vec_length)
            new_image[i:i + tile_y, j:j + tile_x] = new_tile
    return new_image
```

### Adding circular tiles option
Here I modified the **'img_mosaic'** function so that there is an option to create mosaic with circular tiles. This is done by creating a black circular mask for each of the tiles. (I tried making the outside of the mask space filled with the average color of the tile, but this way it is very hard to see the circlular tile shapes, therefore black was choosen)

```python
def img_mosaic(image,db,tile_x,tile_y,vec_length, tile = 'rect'):
    if tile == 'rect':
        new_image = image
        y, x, _ = np.shape(new_image)
        for i in range(0,y,tile_y):
            for j in range(0,x,tile_x):
                old_tile = new_image[i:i + tile_y, j:j + tile_x]
                new_tile = best_tile(db, old_tile, vec_length)
                new_image[i:i + tile_y, j:j + tile_x] = new_tile
    elif tile == 'circ':
        new_image = image
        y, x, _ = np.shape(new_image)
        for i in range(0,y,tile_y):
            for j in range(0,x,tile_x):
                old_tile = new_image[i:i + tile_y, j:j + tile_x]
                new_tile = best_tile(db, old_tile, vec_length)
                lx, ly, _ = new_tile.shape
                X, Y = np.ogrid[0:lx, 0:ly]
                mask = (X - lx/2)**2 + (Y - ly/2)**2 > lx*ly/4
                new_tile[mask] = 0
                new_image[i:i + tile_y, j:j + tile_x] = new_tile
    return new_image
```
